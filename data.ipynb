```python
import cdflib
import pandas as pd
import matplotlib.pyplot as plt

cdf = cdflib.CDF("Data\AL1_ASW91_L2_TH2_20250625_UNP_9999_999999_V02.cdf")

# Extract core variables
time = cdflib.cdfepoch.to_datetime(cdf.varget("epoch_for_cdf_mod"))
energy_center = cdf.varget("energy_center_mod")
flux = cdf.varget("integrated_flux_mod")  # likely shape: [time, energy_bin]

# Create a DataFrame for 1 energy bin (e.g., the middle one)
middle_bin = flux.shape[1] // 2
flux_series = flux[:, middle_bin]

df = pd.DataFrame({
    "time": pd.to_datetime(time),
    "flux_bin_center_keV": [energy_center[middle_bin]] * len(time),
    "flux": flux_series
})

# Plot to explore
plt.figure(figsize=(12, 5))
plt.plot(df["time"], df["flux"])
plt.title("Integrated Flux over Time TH2 (middle energy bin)")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.grid(True)
plt.show()
```

```python
import cdflib
import pandas as pd
import numpy as np

# Load the CDF file
cdf = cdflib.CDF("Data\AL1_ASW91_L2_TH1_20250625_UNP_9999_999999_V02.cdf")

# Variables confirmed to be available in TH1
useful_vars = [
    'epoch_for_cdf_mod',
    'energy_center_mod',
    'energy_uncer',
    'integrated_flux_mod',
    'flux_uncer',
    'integrated_flux_s9_mod',
    'integrated_flux_s10_mod',
    'integrated_flux_s11_mod',
    'spacecraft_xpos',
    'spacecraft_ypos',
    'spacecraft_zpos',
    'sun_angle_tha1'
]

data_dict = {}

for var in useful_vars:
    try:
        data = cdf.varget(var)

        # Convert epoch to readable datetime
        if var == 'epoch_for_cdf_mod':
            data = cdflib.cdfepoch.to_datetime(data)
            data_dict[var] = data
            continue

        # 1D variables (most common)
        if isinstance(data, np.ndarray) and data.ndim == 1:
            data_dict[var] = data

        # 2D variables (just in case)
        elif isinstance(data, np.ndarray) and data.ndim == 2:
            for i in range(data.shape[1]):
                col_name = f"{var}_{i}"
                data_dict[col_name] = data[:, i]

        else:
            print(f"⚠ Skipping {var}: unsupported shape {data.shape}")

    except Exception as e:
        print(f"❌ Error reading {var}: {e}")

# Convert to DataFrame and save
df2 = pd.DataFrame(data_dict)
df2.to_csv("th1_swis.csv", index=False)
print("✅ Saved to: th1_swis.csv | Rows:", df2.shape[0], "| Columns:", df2.shape[1])
df2.head()
```
